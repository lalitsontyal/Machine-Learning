```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
data = pd.read_csv(io.BytesIO(uploaded['polynomial_regression.csv']))
data
```

```python
data.head()
```

```python
#From the dataset we are dropping position and salary
X = data.drop(["Position","Salary"] , axis = 1)
#It holds the value of Salary column from our dataset
y = data["Salary"]
```

```python
#Data Visualization
plt.scatter(X, y , label = "Distribution")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
#Create Polynomial features
poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X)
```

```python
#data split
X_train , X_test , y_train , y_test = train_test_split(X_poly , y, test_size = 0.10 , random_state = 42)
```

```python
#model
model = LinearRegression()
model = model.fit(X_train , y_train)
```

```python
#Prediction
y_pred = model.predict(X_test)
```

```python
#Control
MSE = mean_squared_error(y_test , y_pred)
print("MSE = {}".format(MSE))
```

```python
#Linear Regression
model_linear = LinearRegression()
model_linear = model_linear.fit(X, y)
```

```python
#Prediction
y_pred_linear = model_linear.predict(X)
```

```python
y_pred_all = model.predict(X_poly)
```

```python
#Visualization
plt.scatter(X, y , label = "Distribution" , color = "navy")
plt.plot(X, y_pred_all, label = "Polynomial Regression" , color = "orange" , linewidth = 4)
plt.plot(X, y_pred_linear, label = "Linear Regression" , color = "gray")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
